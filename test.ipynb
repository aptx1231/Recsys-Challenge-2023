```python
import lightgbm as lgb
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# 加载数据
X_train = np.random.rand(1000, 10)
y_train = np.random.randint(0, 3, 1000)

X_test = np.random.rand(100, 10)
y_test = np.random.randint(0, 3, 100)

# 将标签转换为类别型变量
# y_train = y_train.astype('category')
# y_test = y_test.astype('category')

# 构造 Dataset
train_data = lgb.Dataset(X_train, label=y_train)
test_data = lgb.Dataset(X_test, label=y_test)

# 设置参数
params = {
    'objective': 'multiclass',
    'num_class': 3,
    'metric': ['multi_logloss', 'multi_error'] # 设置损失函数和准确率
}

# 训练模型
model = lgb.train(params, train_data, num_boost_round=100)

# 预测
y_pred = model.predict(X_test)
y_pred_label = np.argmax(y_pred, axis=1)

# 计算指标
# loss = model.eval(test_data, metric=params['metric'][0])['multi_logloss'][0]
acc = accuracy_score(y_test, y_pred_label)
precision = precision_score(y_test, y_pred_label, average='macro')
recall = recall_score(y_test, y_pred_label, average='macro')
f1 = f1_score(y_test, y_pred_label, average='macro')

print(f"Accuracy: {acc:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}, F1 Score: {f1:.4f}")
```
